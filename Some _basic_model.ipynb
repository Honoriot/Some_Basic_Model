```python
# The whole package we consider 
import sklearn # he is the grand one
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine # get the data 
from sklearn.model_selection import train_test_split # divide your data hahaha...
# models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
LW = load_wine()
print('Keys: {}'.format(LW.keys()))
```

```python
# Knowing your data is to important
print("Show Data: {}".format(LW['data'][:5]))
print("Show Data shape: {}".format(LW['data'].shape))
```

```python
print("Show Target: {}".format(LW['target']))
print("Show Target shape: {}".format(LW['target'].shape))
```

```python
print("Show target name: {}".format(LW['target_names']))
```

```python
print('its really important yoyoyoy... print yourself')
#print("Show Data details: {}".format(LW['DESCR']))
```

```python
print("Show features names: {}".format(LW['feature_names']))
```

```python
x_index = 0
y_index = 1

formatter = plt.FuncFormatter(lambda i, *args: LW.target_names[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(LW.data[:, x_index], LW.data[:, y_index], c= LW.target)
plt.colorbar(ticks=[0,1,2], format= formatter)

plt.tight_layout()

plt.show()
```

```python
# its time to split your data ...
# in train and test

X_train, X_test, y_train, y_test = train_test_split(LW['data'], LW['target'], random_state = 0)
print('X_train: {}'.format(X_train.shape))
print('y_train: {}'.format(y_train.shape))
```

```python
# Start Modeling
# its KNN bro
KNN = KNeighborsClassifier(n_neighbors = 1)

KNN.fit(X_train, y_train)
pre1 = KNN.predict(X_test)
print("pre1: {}".format(pre1))
print("Accuracy on training set: {:.2f}".format(KNN.score(X_train, y_train)))
print("Accuracy on test set: {:.2f}".format(KNN.score(X_test, y_test)))
```

```python
# Start Modeling
#its Tree bro
tree = DecisionTreeClassifier(max_depth=7, random_state=0)# max_depth is a predicting factor
tree.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
# Start Modeling
#its LogisticRegression bro
logreg = LogisticRegression(C = 10000).fit(X_train, y_train)# C is a predicting factor
print("Training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
```
